```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 6.626e-34  # Planck's constant (Joule seconds)
c = 3.0e8      # Speed of light (m/s)
kB = 1.38e-23  # Boltzmann constant (Joules per Kelvin)
T = 300        # Temperature in Kelvin (room temperature)
frequencies = np.linspace(1e10, 1e14, 1000)  # Define a range of frequencies (adjust as needed)
```

```python
# Planck's law
def planck_law(f):
    return (2.0 * h * f**3) / (c**2 * (np.exp((h * f) / (kB * T)) - 1))

# Rayleigh-Jeans law
def rayleigh_jeans_law(f):
    return (8.0 * np.pi * f**2 * kB * T) / (c**3)

planck_values = [planck_law(f) for f in frequencies]
rayleigh_jeans_values = [rayleigh_jeans_law(f) for f in frequencies]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(frequencies, planck_values, label="Planck's Law")
plt.plot(frequencies, rayleigh_jeans_values, label="Rayleigh-Jeans Law")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Spectral Radiance (W/sr/m^2/Hz)')
plt.title("Planck's Law vs Rayleigh-Jeans Law")
plt.legend()
plt.grid(True)
plt.show()
```
